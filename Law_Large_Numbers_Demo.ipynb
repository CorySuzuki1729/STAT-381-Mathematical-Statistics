```python
# STAT 381: Law of Large Numbers Simulation 
# Author: Cory Suzuki

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Illustrates LLN by showing that the sample mean converges to the
# population mean as sample size increases.

n_trials = 1000
expected_value = 3.5

# 2. Simulate rolling a die (random integers between 1 and 6)
rolls = np.random.randint(1, 7, size=n_trials)

# 3. Calculate cumulative averages
# np.cumsum provides the running sum; divide by trials count [1, 2, ..., n]
cumulative_avg = np.cumsum(rolls) / np.arange(1, n_trials + 1)

# 4. Visualization
plt.figure(figsize=(10, 6))
plt.plot(cumulative_avg, label="Sample Mean (Observed)")
plt.axhline(y=expected_value, color='r', linestyle='--', label=f"True Mean ({expected_value})")
plt.title("Law of Large Numbers: Die Roll Simulation")
plt.xlabel("Number of Trials")
plt.ylabel("Cumulative Average")
plt.legend()
```
