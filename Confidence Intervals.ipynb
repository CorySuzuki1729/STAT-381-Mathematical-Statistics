```python
# STAT 381: Confidence Intervals
# Author: Cory Suzuki

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# 1 sample proportion z interval

n = 100
x = 40
p_hat = x / n
sig_lvl = 0.05
z_crit = stats.norm.ppf(1- sig_lvl/2)
se_hat = np.sqrt((p_hat * (1 - p_hat)) / n )
margin_of_error = z_crit * se_hat
CI_lower = p_hat - margin_of_error
CI_upper = p_hat + margin_of_error
print(f"({CI_lower}, {CI_upper})")

# 2 sample proportion z interval

n1 = 150
x1 = 30
p1_hat = x1 / n1
n2 = 200
x2 = 50
p2_hat = x2 / n2
sig_lvl = 0.05
z_crit_2sp = stats.norm.ppf(1 - sig_lvl/2)
se_hat_2sp = np.sqrt((p1_hat * (1 - p1_hat) / n1) + (p2_hat * (1 -  p2_hat) / n2))
margin_of_error_2sp = z_crit_2sp ** se_hat_2sp
CI_lower_2sp = (p1_hat - p2_hat) - margin_of_error_2sp
CI_upper_2sp = (p1_hat - p2_hat) + margin_of_error_2sp
print(f"({CI_lower_2sp}, {CI_upper_2sp})")

# 1 sample mean z interval (known sigma)

n = 36
x_bar = 80
sigma = 12
sig_lvl = 0.05
z_crit_zm = stats.norm.ppf(1 - sig_lvl/2)
se_hat_zm = sigma / np.sqrt(n)
margin_of_error_zm = z_crit_zm ** se_hat_zm
CI_lower_zm = x_bar - margin_of_error_zm
CI_upper_zm = x_bar + margin_of_error_zm
print(f"({CI_lower_zm}, {CI_upper_zm})")

# 1 sample mean t interval (unknown sigma)

n = 25
x_bar = 100
s = 15
sig_lvl = 0.05
df = n - 1
t_crit_tm = stats.t.ppf(1 - sig_lvl/2, df)
se_hat_tm = s / np.sqrt(n)
margin_of_error_tm = t_crit_tm * se_hat_tm
CI_lower_tm = x_bar - margin_of_error_tm
CI_upper_tm = x_bar + margin_of_error_tm
print(f"({CI_lower_tm}, {CI_upper_tm})")

# 2 sample mean z interval (known sigmas)

n1 = 30
x1_bar = 75
sigma1 = 10
n2 = 35
x2_bar = 70
sigma2 =  12
sig_lvl = 0.05
z_crit_2sm = stats.norm.ppf(1 - sig_lvl/2)
se_hat_2sm = np.sqrt((sigma1 ** 2 / n1) + (sigma2 ** 2 / n2))
margin_of_error_2sm = z_crit_2sm * se_hat_2sm
CI_lower_2sm = (x1_bar - x2_bar) - margin_of_error_2sm
CI_upper_2sm = (x1_bar - x2_bar) + margin_of_error_2sm
print(f"({CI_lower_2sm}, {CI_upper_2sm})")

# 2 sample mean t interval (unknown sigmas)

n1 = 30
x1_bar = 75
s1 = 10
n2 = 35
x2_bar = 70
s2 =  12
sig_lvl = 0.05
df = n1 + n2 - 2
t_crit_2m = stats.t.ppf(1 - sig_lvl/2, df)
se_hat_2smt = np.sqrt((n1 - 1) * s1**2 + (n2 - 1) * s2**2) / (df)
margin_of_error_2smt = t_crit_2m * se_hat_2smt * np.sqrt(1/n1 + 1/n2)
CI_lower_2smt = (x1_bar - x2_bar) - margin_of_error_2smt
CI_upper_2smt = (x1_bar - x2_bar) + margin_of_error_2smt
print(f"({CI_lower_2smt}, {CI_upper_2smt})")

# 1 sample CI for variance

n = 6
sig_lvl = 0.10
sig_lvl_lower = sig_lvl / 2
sig_lvl_upper = 1 - (sig_lvl / 2)
df = n - 1
sample_mean = 85.73
sample_var = 0.5027
chi2_crit_lower = stats.chi2.ppf(sig_lvl_lower, df)
chi2_crit_upper = stats.chi2.ppf(sig_lvl_upper, df)
CI_lower_var = (df * sample_var) / chi2_crit_upper
CI_upper_var = (df * sample_var) / chi2_crit_lower
print(f"({CI_lower_var}, {CI_upper_var})")
```
