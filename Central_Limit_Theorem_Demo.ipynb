```python
# STAT 381: Central Limit Theorem Simulation 
# Author: Cory Suzuki

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Choosing different sample sizes ranging from small to large to 
# illustrate the CLT. This example samples from an exponential distribution.

sample_sizes = [1, 5, 10, 100, 1000, 100000]
all_sample_means = []

np.random.seed(42)

for size in sample_sizes:
    sample_means = [np.mean(np.random.exponential(scale=2.0, size=size))
                    for _ in range(1000)]
    all_sample_means.append(sample_means)

fig, axes = plt.subplots(3, 2, figsize=(10, 8))
for ax, means, size in zip(axes.flatten(), all_sample_means, sample_sizes):
    ax.hist(means, bins=20, density=True, alpha=0.75,
            color='pink', edgecolor='black')
    ax.set_title(f'Sample Size = {size}')
    ax.set_xlabel("Sample Mean")
    ax.set_ylabel("Density")
    ax.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
